```python
df = pd.read_csv("assets/data/small_mnist.csv")
df.head()
```

```python
# check max array value
df.iloc[1:].values.max()
```

```python
from sklearn.model_selection import train_test_split

# scale dengan 255
X = df.drop(columns="label").values / 255
y = df.label.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

<h1> Visualize Random Data

```python
N = len(X_train)
idx = np.random.choice(N)
img, label = X_train[idx], y_train[idx]

plt.imshow(img.reshape(28, 28), cmap='gray')
plt.title('Label: {}'.format(label))
plt.axis('off')
```

<h1> Training

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf_params = {
    'n_estimators': [100, 200, 300],  # Jumlah pohon keputusan dalam ensemble
    'max_depth': [None, 10, 20],  # Maksimum kedalaman pohon keputusan
    'min_samples_split': [2, 5, 10],  # Jumlah minimum sampel yang diperlukan untuk membagi simpul
    'min_samples_leaf': [1, 2, 4],  # Jumlah minimum sampel yang diperlukan untuk simpul daun
    'max_features': ['auto', 'sqrt'],  # Jumlah fitur yang dipertimbangkan saat mencari pemisahan terbaik
    'bootstrap': [True, False],  # Penggunaan bootstrap sampling
    'random_state': [42]  # Seed untuk keacakan yang dapat direproduksi
}
```

```python
from jcopml.tuning import random_search_params as rsp
```

```python
pipeline = Pipeline([
    ('algo', RandomForestClassifier(n_jobs=-1, random_state=42))
])

model = RandomizedSearchCV(pipeline, rsp.rf_params, n_iter=50, cv=3, n_jobs=-1, verbose=1 ,random_state=42)
model.fit(X_train, y_train)

print('Best score: ', model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

<h1> Visualize Prediction

```python
idx = np.random.choice(len(X_test), size=36, replace=False)
images, labels = X_test[idx], y_test[idx]
preds = model.predict(images)

fig, axes = plt.subplots(6, 6, figsize=(12, 12))
for img, label, pred, ax in zip(images, labels, preds, axes.flatten()):
    font = {'color': 'r'} if label != pred else {'color': 'g'}
    ax.imshow(img.reshape(28, 28), cmap='gray')
    ax.set_title(f"Label: {label}, Pred: {pred}", fontdict=font)
    ax.axis('off')
```
