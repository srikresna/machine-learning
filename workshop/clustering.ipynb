```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering, SpectralClustering

from luwiji.cluster import demo
import warnings
warnings.filterwarnings("ignore")
```

```python
X1, y1 = demo.blob_data()
X2, y2 = demo.moon_data()
X3, y3 = demo.circle_data()
```

```python
X = X1
plt.scatter(X[:, 0], X[:, 1], s=10)
plt.axis('equal')
```

```python
X = X2
plt.scatter(X[:, 0], X[:, 1], s=10)
plt.axis('equal')
```

```python
X = X3
plt.scatter(X[:, 0], X[:, 1], s=10)
plt.axis('equal')
```

</h3> K-means clustering

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X1)
member = kmeans.predict(X)
```

```python
member
```

```python
centroid = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(12, 9))
plt.scatter(X1[:, 0], X1[:, 1], s=10, c=y1) # kita pakai label untuk cek hasilnya
plt.scatter(centroid[:, 0], centroid[:, 1], c='r', s=200, edgecolors='k', marker='X')
plt.axis('equal')
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X2)
member = kmeans.predict(X2)
centroid = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(12, 9))
# plt.scatter(X2[:, 0], X2[:, 1], s=10, c=y1) # kita pakai label untuk cek hasilnya
plt.scatter(X2[:, 0], X2[:, 1], s=10, c=member)
plt.scatter(centroid[:, 0], centroid[:, 1], c='w', s=200, edgecolors='k', marker='X')
plt.axis('equal')
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X3)
member = kmeans.predict(X3)
centroid = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(12, 9))
# plt.scatter(X2[:, 0], X2[:, 1], s=10, c=y1) # kita pakai label untuk cek hasilnya
plt.scatter(X3[:, 0], X3[:, 1], s=10, c=member)
plt.scatter(centroid[:, 0], centroid[:, 1], c='w', s=200, edgecolors='k', marker='X')
plt.axis('equal')
```

Menentukan K (n_cluster) tanpa label  inertia / Elbow analysis

```python
kmeans.inertia_
```

```python
X = X1
score = []

for k in range(2, 25):
    kmeans = KMeans(n_clusters=k)
    member = kmeans.fit_predict(X)
    score.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(2, 25), score, "r-") # warna merah
plt.ylabel("Inertia")
plt.xlabel("k")
```

Menentukan K (n_cluster) dan ada label  Silhoutte Score

```python
from sklearn.metrics import silhouette_score
```

```python
X = X1
score = []

for k in range(2, 25):
    kmeans = KMeans(n_clusters=k)
    member = kmeans.fit_predict(X)

    siluet = silhouette_score(X, member, metric = 'euclidean')
    score.append(siluet)

plt.figure(figsize=(10, 5))
plt.plot(range(2, 25), score, "r-") # warna merah
plt.ylabel("Silhouette Score")
plt.xlabel("k")
```

```python
# show hierarchy
from scipy.cluster.hierarchy import dendrogram, ward, linkage, fcluster

# demo
X, y = demo.blob_data(30)
plt.scatter(X[:, 0], X[:, 1], s=10)

# linkage matrix
linkage_matrix = linkage(X, method='ward', metric='euclidean')

# plot dendrogram
plt.figure(figsize=(12, 9))
dendrogram(linkage_matrix);

# cut dendrogram
# plt.figure(figsize=(12, 9))
```
