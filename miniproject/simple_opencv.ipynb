```python
# import library
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
def plt_imshow(title, image):
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
# masukkan nama file yang ingin diproses
args = {
	"image": "data/afif.jpg",
}
```

```python
# load gambar menggunakan fungsi cv2.imread()
image = cv2.imread(args["image"])
(h, w, c) = image.shape[:3]

# melihat detil gambar
print("width: {} pixels".format(image.shape[1]))
print("height: {}  pixels".format(image.shape[0]))
print("channels: {}".format(image.shape[2]))
```

```python
# tampilkan gambar
plt_imshow("Gambar", image)
```

```python
# Tampilkan piksel pada posisi 0,0
(b, g, r) = image[0, 0]
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r, g, b))

# Tampilkan piksel pada lokasi :  x=50, y=20
(b, g, r) = image[20, 50]
print("Pixel at (50, 20) - Red: {}, Green: {}, Blue: {}".format(r, g, b))

# update piksel pada (50, 20) dengan warna merah
image[20, 50] = (0, 0, 255)
(b, g, r) = image[20, 50]
print("Pixel at (50, 20) - Red: {}, Green: {}, Blue: {}".format(r, g, b))
```

```python
# Hitung koordinat tengah pada gambar
(cX, cY) = (w // 2, h // 2)

# Lakukan slicing pada gambar
# Tampilkan gambar di bagian kiri atas
tl = image[0:cY, 0:cX]
plt_imshow("Pojok Kiri Atas", tl)
```

```python
# Lakukan hal yang sama untuk ketiga potongan lainnya
tr = image[0:cY, cX:w]
br = image[cY:h, cX:w]
bl = image[cY:h, 0:cX]
plt_imshow("Pojok Kanan Atas", tr)
plt_imshow("Pojok Kanan Bawah", br)
plt_imshow("Pojok Kiri Bawah", bl)
```

```python
# Ubah piksel gambar di pojok kiri atas dengan warna hijau
image[0:cY, 0:cX] = (188, 198, 252)

# Tampilkan gambar
plt_imshow("Updated", image)
```

## Menggambar Dengan OpenCV

```python
# Inisiasi canvas kosong dengan ukuran 300 x 300 pixel dan 3 channel
canvas = np.zeros((300, 300, 3), dtype="uint8")
```

```python
plt_imshow("Canvas", canvas)
```

```python
# Gambar garis hijau secara diagonal (kiri atas - kanan bawah)
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)
plt_imshow("Canvas", canvas)
```

```python
# Gambar garis merah 3x lebih tebal dengan arah diagonal sebaliknya
red = (0, 0, 255)
cv2.line(canvas, (300, 0), (0, 300), red, 3)
plt_imshow("Canvas", canvas)
```

```python
# Gambar persegi ukuran 50x50 piksel berwarna hijau , dari (10,10) sampai (60,60)
cv2.rectangle(canvas, (10, 10), (60, 60), green)
plt_imshow("Canvas", canvas)
```

```python
# Gambar persegi lainnya yang lebih tebal dengan warna merah
cv2.rectangle(canvas, (50, 200), (200, 225), red, 5)
plt_imshow("Canvas", canvas)
```

```python
# Gambar persegi lainnya, warna biru dan berisi
blue = (255, 0, 0)
cv2.rectangle(canvas, (200, 50), (225, 125), blue, -1)
plt_imshow("Canvas", canvas)
```

```python
# inisiasi ulang canvas kosong dan hitung titik tengahnya
canvas = np.zeros((300, 300, 3), dtype="uint8")
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2)
white = (255, 255, 255)

# Looping dan buat lingkaran
for r in range(0, 175, 25):
	cv2.circle(canvas, (centerX, centerY), r, white, 4)

plt_imshow("Canvas", canvas)
```

```python
# Inisiasi ulang canvas kosong
canvas = np.zeros((300, 300, 3), dtype="uint8")

# buat gambar "abstrak"
for i in range(0, 25):
	radius = np.random.randint(5, high=200)
	color = np.random.randint(0, high=256, size=(3,)).tolist()
	pt = np.random.randint(0, high=300, size=(2,))

	cv2.circle(canvas, tuple(pt), radius, color, -1)

plt_imshow("Canvas", canvas)
```

```python
cv2.imwrite("/content/gdrive/My Drive/hasil_workshop/masterpiece.jpg", canvas)
```
