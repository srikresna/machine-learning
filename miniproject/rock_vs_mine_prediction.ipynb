```python
#import library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data = pd.read_csv("/content/Copy of sonar data.csv", header = None)
```

```python
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe() #statistical measure of the data
```

```python
print(f"Count M represent Mine & R represent Rock :\n{sonar_data[60].value_counts()}")
```

```python
sonar_data.groupby(60).mean()
```

```python
X = sonar_data.drop(columns = 60, axis = 1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

```python
#split dataframe into training data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state = 1) 
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#model training using logistik regresion

model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
#evaluate the model

X_train_prediction = model.predict(X_train)
X_test_prediction = model.predict(X_test)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print(f"Accuracy of training data : {training_data_accuracy}")
print(f"Accuracy of test data : {test_data_accuracy}")
```
