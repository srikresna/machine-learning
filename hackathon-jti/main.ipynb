```python

kecamatan_dict = {
    "Blimbing": ["Arjosari", "Balearjosari", "Blimbing", "Bunulrejo", "Jodipan", "Kesatrian", "Pandanwangi", "Polehan", "Polowijen", "Purwantoro", "Purwodadi"],
    "Kedungkandang": ["Arjowinangun", "Bumiayu", "Buring", "Cemorokandang", "Kedungkandang", "Kotalama", "Lesanpuro", "Madyopuro", "Mergosono", "Sawojajar", "Tlogowaru", "Wonokoyo"],
    "Klojen": ["Bareng", "Gadingasri", "Kasin", "Kauman", "Kiduldalem", "Klojen", "Oro-Oro Dowo", "Penanggungan", "Rampal Celaket", "Samaan", "Sukoharjo"],
    "Lowokwaru": ["Dinoyo", "Jatimulyo", "Ketawanggede", "Lowokwaru", "Merjosari", "Mojolangu", "Sumbersari", "Tasikmadu", "Tlogomas", "Tulusrejo", "Tunggulwulung", "Tunjungsekar"],
    "Sukun": ["Bakalankrajan", "Bandulan", "Bandungrejosari", "Ciptomulyo", "Gadang", "Karangbesuki", "Kebonsari", "Mulyorejo", "Pisangcandi", "Sukun", "Tanjungrejo"]
}

kelurahan_to_kecamatan = {kelurahan: kecamatan for kecamatan, kelurahans in kecamatan_dict.items() for kelurahan in kelurahans}

def extract_kelurahan_from_address(address):
    if isinstance(address, str):
        for kelurahan in kelurahan_to_kecamatan:
            if re.search(r'\b' + re.escape(kelurahan) + r'\b', address, re.IGNORECASE):
                return kelurahan
    return None

def extract_kecamatan_from_address(address):
    kelurahan = extract_kelurahan_from_address(address)
    if kelurahan:
        return kelurahan_to_kecamatan[kelurahan]
    if isinstance(address, str):
        for kecamatan in kecamatan_dict:
            if re.search(r'\b' + re.escape(kecamatan) + r'\b', address, re.IGNORECASE):
                return kecamatan
    return None

df['Kecamatan'] = df['Address'].apply(extract_kecamatan_from_address)
df['Kelurahan'] = df['Address'].apply(extract_kelurahan_from_address)
```

```python
def extract_popular_times(popular_time_str):
    pattern = r'(\d+)% busy at (\d+\u202f?[APM]+)'
    
    matches = re.findall(pattern, str(popular_time_str))  # Pastikan input adalah string
    
    if matches:
        popular_df = pd.DataFrame(matches, columns=['Percentage', 'Time'])
        popular_df['Time'] = popular_df['Time'].str.replace('\u202f', '', regex=True)
        popular_df['Percentage'] = popular_df['Percentage'].astype(int)
        return popular_df
    else:
        return pd.DataFrame(columns=['Percentage', 'Time'])  # Jika tidak ada data

def get_busiest_time(popular_times):
    if not popular_times.empty:
        busiest_time = popular_times.loc[popular_times['Percentage'].idxmax()]
        return busiest_time['Time']
    return None

days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
for day in days_of_week:
    column_name = f'Popular_times_{day}'
    df[f'Busiest_time_{day}'] = df[column_name].apply(lambda x: get_busiest_time(extract_popular_times(x)))

df[[f'Busiest_time_{day}' for day in days_of_week]].head()
```
